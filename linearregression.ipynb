```python
#calculating VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data=X

##calculating vif for every column
VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF
```

```python
VIF[VIF==VIF.max()].index[0]
```

```python
def MC_remover(data):
    vif=pd.Series([variance_inflation_factor(data.values,i) for i in range(data.shape[1])],index=data.columns)
    if vif.max()>5:
        print(vif[vif==vif.max()].index[0],'has been removed')
        data=data.drop(columns=[vif[vif==vif.max()].index[0]])
        return data
    else:
        print('no multicollinearity present anymore')
        return data
```

```python
for i in range(7):
    vif_data=MC_remover(vif_data)

vif_data.head()
```

```python
def adj_r2(data_x,data_y,predictions):
#x are independent variables,y is the target variable
    from sklearn.metrics import r2_score
    R=r2_score(data_y,predictions)
    n=len(data_y)
    m=len(data_x.columns)
    adj_R=1-((1-R)*(n-1))/(n-m-1)
    return adj_R
```

```python
adj_r2(x_test,y_test,predictions)
```

```python
def return_r2(X,Y,model):
    '''
    X:independent var
    Y:deoendent var
    model:predictive model
    returns adjusted R2 corresponding tio the model
    '''
    model.fit(X,Y)
    pred=model.predict(X)
    m=len(X.columns)
    r2=adj_r2(m,Y,pred)
    return r2
```
