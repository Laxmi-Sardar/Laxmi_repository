```python
##froward selection
def adj_r2( m ,data_y, predictions):
    #calculate adjusted_r2 value
  '''
  m : number of independent variables
  data_y : target variable
  predictions : prediction of the model to be evaluated

  Returns: Adjusted R2
  '''

  from sklearn.metrics import r2_score
  R = r2_score(data_y, predictions)
  n = len(data_y)
  adj_R = 1 - ((1-R)*(n-1))/(n-m-1)
  return adj_R
    
def return_r2(X, Y, model):
  '''
  X : independent variables
  Y : target variable
  model : predictive model

  Returns: Adjusted R2 corresponding to the model
  '''
  model.fit(X, Y)
  pred = model.predict(X)
  m = len(X.columns)
  r2 = adj_r2(m, Y, pred)
  return r2
    
def forward_selection(model, X, Y, max_features):
  '''
  X : independent variables
  Y : target variable
  model : any predictive model
  max_features : selects best feature upto max_features value

  Returns: 
  Rmax : Adjusted R2 of the final model
  history : list of Rmax at every new independent variable added to the model
  x[f_col] : best features upto max_features
  '''
  
  f_col = []#store var being selected at each dtep of forward selection
  history = []#store adj r2 values whenever new feature is selected
  
  for i in range(max_features):#loop until amximum number of featurs are selected

    f_col.append('blank')#updates when we encounter a element for  best feature
    tmp = f_col[:]
    Rmax = -1 #keeps track that adding which variable causes maximum adjuted r2 result

    for var in X.columns:
      tmp[-1] = var#last element of tmp in selected new col to travel if yess update r2
      r2 = return_r2( X[tmp] , Y, model)
      if r2 >= Rmax:#checking weather adding this var leads to highest r2 value 
        Rmax = r2
        f_col[-1] = var# the new variable with highest adjr2 value is retained
        elim = var
  
    print("Selected",f_col[-1],"with new r2 =",Rmax)
    history.append(Rmax)

  return Rmax, history, X[f_col]
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

score, history, sel_data = forward_selection(lr, x_train, y_train, 5)
```

```python
#backward elimination
def adj_r2( m=0, data_y=0, predictions=0):
  '''
  m : number of independent variables
  data_y : target variable
  predictions : prediction of the model to be evaluated

  Returns: Adjusted R2
  '''
  from sklearn.metrics import r2_score
  R = r2_score(data_y, predictions)
  n = len(data_y)
  adj_R = 1 - ((1-R)*(n-1))/(n-m-1)
  return adj_R
    
def return_r2(X, Y, model):
  '''
  X : independent variables
  Y : target variable
  model : predictive model

  Returns: Adjusted R2 corresponding to the model
  '''
  model.fit(X, Y)
  pred = model.predict(X)
  m = len(X.columns)
  r2 = adj_r2(m, Y, pred)
  return r2
    
def backward_elimination( model, X, Y, min_features):
  '''
  X : independent variables
  Y : target variable
  model : predictive model
  min_features : eliminate worst features until only min_features are left

  Returns: 
  Rmax : Adjusted R2 of the final model
  history : list of Rmax at every new independent variable added to the model
  x[f_col] : best features upto max_features
  '''
  elim_data = X[:]#contains all cols in x_train
  r2_history = []#keep track of adiR2 when a feature is eliminated
  base_r2 = return_r2(elim_data, Y, model)#the baseline adjr2 score when independent variables are used to build A LR
  print("base_r2 is", base_r2) 


  for i in range(X.shape[1] - min_features):#run loop until minimu features are accquired
  
    difference = -100

    for var in elim_data.columns:
      tmp = elim_data.drop(columns = [var])
      r2 = return_r2(tmp, Y, model)

      if (r2 - base_r2) > difference:#dropping the feature  had maximum desired effect
        difference = r2 - base_r2
        eliminate = var
        new_r2 = r2
  
    elim_data.drop(columns = [eliminate], inplace = True)
    r2_history.append(new_r2)
    print(eliminate,"Eliminated : New adj_r2 = ",new_r2,)
    base_r2 = new_r2

  return base_r2, r2_history, elim_data
```
